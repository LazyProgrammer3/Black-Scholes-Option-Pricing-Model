# <span style="color:red">1- Introduction Théorique aux Options</span>

## Qu'est-ce qu'une Option ?

Une option est un contrat financier qui donne à son détenteur le droit, mais non l'obligation, d'acheter ou de vendre un actif sous-jacent à un prix spécifié, appelé prix d'exercice, avant ou à une date d'expiration déterminée. Les options sont des instruments dérivés, ce qui signifie que leur valeur dérive de la valeur d'un autre actif, appelé actif sous-jacent.

## Types d'Options

Il existe deux principaux types d'options :

1. **Option d'achat (Call)** : Donne à l'acheteur le droit d'acheter l'actif sous-jacent à un prix d'exercice spécifié.
2. **Option de vente (Put)** : Donne à l'acheteur le droit de vendre l'actif sous-jacent à un prix d'exercice spécifié.


## Terminologie Clé

- **Prix d'exercice (Strike Price)** : Le prix auquel l'actif sous-jacent peut être acheté (pour un call) ou vendu (pour un put).
- **Date d'expiration (Expiration Date)** : La date à laquelle l'option expire. Après cette date, l'option n'a plus de valeur.
- **Prime de l'option (Option Premium)** : Le prix que l'acheteur de l'option paie pour acquérir ce droit.
- **Actif sous-jacent (Underlying Asset)** : L'actif sur lequel porte l'option, tel qu'une action, une matière première, une devise, etc.
- **Américaines vs Européennes** : Les options américaines peuvent être exercées à tout moment avant la date d'expiration, tandis que les options européennes ne peuvent être exercées qu'à la date d'expiration.

## Facteurs Influant sur le Prix des Options

Le prix des options, ou prime, est influencé par plusieurs facteurs :

1. **Le prix de l'actif sous-jacent** : Plus le prix de l'actif sous-jacent est élevé par rapport au prix d'exercice, plus la valeur d'une option d'achat augmente, et inversement pour une option de vente.
2. **Le prix d'exercice** : Un prix d'exercice plus élevé rend une option d'achat moins attrayante, tandis qu'un prix d'exercice plus bas rend une option de vente moins attrayante.
3. **Le temps restant jusqu'à l'expiration** : Plus la date d'expiration est éloignée, plus l'option a de la valeur en raison de l'incertitude plus grande.
4. **La volatilité de l'actif sous-jacent** : Une plus grande volatilité augmente la valeur des options, car il y a une plus grande probabilité que l'option devienne profitable.
5. **Le taux d'intérêt sans risque** : Un taux d'intérêt sans risque plus élevé augmente la valeur des options d'achat et diminue la valeur des options de vente.
6. **Les dividendes attendus** : Les dividendes à venir peuvent diminuer la valeur des options d'achat et augmenter la valeur des options de vente.


## Utilisation des Options

Les options sont utilisées pour diverses stratégies financières, incluant :

1. **Couverture (Hedging)** : Les investisseurs utilisent des options pour protéger leurs portefeuilles contre les mouvements défavorables du marché.
2. **Spéculation** : Les traders spéculent sur les mouvements futurs des prix des actifs sous-jacents pour réaliser des profits.
3. **Arbitrage** : Les traders exploitent les différences de prix entre les marchés pour réaliser des profits sans risque.


# <span style="color:red">2- Modèle de Black-Scholes</span>

## <span style="color:black">Introduction</span>

Le modèle de Black-Scholes, également connu sous le nom de modèle Black-Scholes-Merton, est l'un des modèles les plus célèbres pour le pricing des options. Il a été développé par Fischer Black, Myron Scholes et Robert Merton en 1973. Ce modèle permet de déterminer le prix théorique des options européennes en fonction de plusieurs variables.

## <span style="color:black">Hypothèses du Modèle</span>

Pour utiliser le modèle de Black-Scholes, plusieurs hypothèses doivent être respectées :

1. Le marché est efficient, ce qui signifie que les prix des actifs suivent un mouvement brownien géométrique.
2. Il n'y a pas de coûts de transaction ou de taxes.
3. Les taux d'intérêt sont constants et connus.
4. Les dividendes ne sont pas pris en compte dans la version de base du modèle.
5. Les options sont de type européen, ce qui signifie qu'elles ne peuvent être exercées qu'à la date d'expiration.

Le modèle de Black-Scholes stipule que les prix des actifs en fonction du temps, \( S(t) \), suivent un mouvement brownien géométrique (GBM) :

$$
dS = S \mu \, dt + S \sigma \, dW_t \,.
$$

Considérons maintenant une option \( $V$), qui sera une fonction de \( S \) et \( t \). Le processus suivi par cette option sera décrit par :
$$ dV(S,t) = \frac{\partial V}{\partial t} dt + \frac{\partial V}{\partial S} dS + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} dt $$

$$
dV(S,t) = \left( \frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} + \frac{1}{2}\sigma^{2}S^{2}\frac{\partial^{2}V}{\partial S^{2}} \right) dt + \frac{\partial V}{\partial S} \sigma S \, dW_t \,.
$$


Considérons le portefeuille suivant $\Pi$, composé de l'option $V$ et d'une certaine quantité ($\Delta$) de l'actif sous-jacent \(S\) :

$$
\Pi = V(S,t) - \Delta S\,.
$$

Alors, ce portefeuille suivra le processus :
$$
d\Pi = dV(S,t) - \Delta \ dS
$$

Nous avons alors défini (delta hedging):
$$
\Delta = \partial V / \partial S\,.
$$
alors :
$$ d\Pi = \left( \frac{\partial V}{\partial t} + \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} \right) dt  $$


Nous remarquons que, dans l'équation pour \( $d\Pi$), la partie stochastique du processus \( $W_t$) n'est plus présente. Cela élimine le risque du portefeuille, qui gagnera donc un taux sans risque \( r \) :

$$
d\Pi = r \Pi dt\,.
(AOA)
$$

En combinant ces équations pour \( $ d\Pi $\), nous obtenons :


$$
\frac{\partial V}{ \partial t} + r\,S\frac{\partial V}{\partial S} +\frac{1}{2}\sigma^{2}S^{2}\frac{\partial^{2}V}{\partial S^{2}} = r\,V\,,
$$
ce qui correspond à l'équation de Black-Scholes.


Le résultat indiquant que le prix d’un call de payoff \($ h(S_T) $\) dans le modèle de Black-Scholes, donné par :

$
v(t, x) := e^{-r(T - t)} \mathbb{E}^{\mathbb{Q}}[h(S_T) \mid S_t = x], \quad (t, x) \in [0, T] \times \mathbb{R}^d,
$

est solution de l'EDP précédent. Ce résultat est très important et est connu sous le nom plus général de **formule de Feynman-Kac**. 

Une espérance conditionnelle sur un processus Markovien peut se réécrire comme solution d’une EDP, créant ainsi des liens entre le monde déterministe et le monde probabiliste.


 La solution de l'équation de Black-Scholes pour une option d'achat européenne \( $C(S,t)$ \), donnée la maturité \( T \) et le prix d'exercice \( K \), est donnée par :

<div style="text-align: center;">
\[ C = S_0 N(d_1) - K e^{-rT} N(d_2) \]
</div>

La formule de parité Call-Put s'écrit :

<div style="text-align: center;">
\[C_t - P_t = S_t - K e^{-r(T - t)}, \quad t \in [0, T]\]
</div>

et pour le prix d'un put européen :

<div style="text-align: center;">
\[ P = K e^{-rT} N(-d_2) - S_0 N(-d_1) \]
</div>


où :
- \($ C $\) est le prix de l'option d'achat (call).
- \($ P $\) est le prix de l'option de vente (put).
- \($ S_0$ \) est le prix actuel de l'actif sous-jacent.
- \($ K $) est le prix d'exercice de l'option.
- \($ T $) est le temps jusqu'à l'expiration (en années).
- \($ r $) est le taux d'intérêt sans risque.
- \( $N(.)$ \) est la fonction de répartition de la loi normale centrée réduite.

- \( $d_1 $\) et \($ d_2 $\) sont définis comme :

<div style="text-align: center;">
\[ d_1 = \frac{\ln(S_0 / K) + (r + \sigma^2 / 2) T}{\sigma \sqrt{T}} \]
   </div>

<div style="text-align: center;">
\[ d_2 = d_1 - \sigma \sqrt{T} = \frac{\ln(S_0/K) + \left(r - \frac{\sigma^{2}}{2}\right)\left(T\right)}{\sigma\sqrt{T}}\,\]
</div>

<div style="text-align: center;">
\[ N(d) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{d} e^{-\frac{1}{2} x^2} \, dx \]
 </div>

 # les Grecques (Sensibilités)
Les sensiblités du prix d’une option par rapport à ses différents paramêtres sont appelés
les Grecques :

## 1.) Delta : Δ

La delta d'une option/portefeuille est la sensibilité du prix par rapport à la valeur actuelle du sous-jacent 

$$
 \Delta = \frac{\partial V}{\partial S} \,
$$

 la delta d'un portefeuille est la somme des deltas de toutes les positions individuelles

→ elle est importante lorsqu'on traite de la couverture delta
Nous pouvons éliminer le risque si nous nous assurons que la delta est égale à la dérivée !!!
vée !!!

## 2.) Gamma : Γ

La gamma d'une option/portefeuille est la dérivée seconde de la position par rapport au sous-jacent
$$
 \Gamma = \frac{\partial^2 V}{\partial^2 S} \,
$$
 le Gamma Γ est la sensibilité du delta (Δ) par rapport à la valeur actuelle du sous-jacent. 




### 3.) Thêta : θ

Le thêta est le taux de variation (la sensibilité )  du prix de l'option avec le temps :
$$
θ = \frac{\partial V}{\partial t} \,
$$
### 4.) Véga :
– Le Vega est la sensibilité du prix par rapport à la volatilité
$$
 \frac{\partial V}{\partial σ} \,
$$
– Certains modèles dépendent fortement de la volatilité, il peut donc être important de surveiller le changement en fonction de la volatilité :


# 5.) Rho ρ :
Le Rho ρ est la sensibilité du prix par rapport au taux d’intérêt r :
$$
ρ = \frac{\partial V}{\partial r} \,
$$

## a) Valeurs des Grecques pour un Call à t = 0

### Delta (∆)
$$
 \Delta = N(d1) > 0 
$$
### Gamma (Γ)
$$
 \Gamma = \frac{1}{x\sigma\sqrt{T}} N'(d1) > 0

$$
 \rho = TKe^{-rT} N(d2) > 0

$$
 \text{Vega} = x\sqrt{T} N'(d1) > 0

### Delta (∆)
$$
\Delta = -N(-d1) > 0 
$$
### Gamma (Γ)
$$
 \Gamma = \frac{1}{x\sigma\sqrt{T}} N'(d1) > 0 
$$
### Rho (ρ)
$$
\rho = TKe^{-rT} (N(d2) - 1) < 0 
$$

$$
 \theta = \frac{x\sigma}{2\sqrt{T}} N'(d1) + Kr e^{-rT} (N(d2) - 1) 
$$
### Vega
$$
 \text{Vega} = x\sqrt{T} N'(d1) > 0 
$$

```python
from math import log, sqrt, exp
from scipy.stats import norm

def BSCall(spot, time, strike, expiry, vol, rate):
    """
    Calcule le prix d'un call selon le modèle de Black-Scholes.

    Paramètres
    ----------
    spot : float
        Le prix spot de l'actif sous-jacent.
    time : float
        Le temps actuel, en années, jusqu'à l'échéance de l'option.
    strike : float
        Le prix d'exercice du call.
    expiry : float
        La date d'expiration du call, en années.
    vol : float
        La volatilité implicite utilisée pour évaluer le call (en pourcentage).
    rate : float
        Le taux d'intérêt sans risque utilisé dans le modèle (en pourcentage).

    Retours
    -------
    float
        Le prix du call selon le modèle de Black-Scholes.
    """
    
    # Convertir volatilité et taux d'intérêt en fractions
    vol /= 100
    rate /= 100
    
    # Calcul des paramètres d1 et d2
    d1 = (log(spot / strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    d2 = d1 - vol * sqrt(expiry - time)
    
    # Calcul du prix du call
    call_price = spot * norm.cdf(d1) - strike * exp(-rate * (expiry - time)) * norm.cdf(d2)
    
    return call_price
```

```python
from math import log, sqrt, exp
from scipy.stats import norm

def BSPut(spot, time, strike, expiry, vol, rate):
    """
    Calcule le prix d'un put selon le modèle de Black-Scholes.

    Paramètres
    ----------
    spot : float
        Le prix spot de l'actif sous-jacent.
    time : float
        Le temps actuel, en années, jusqu'à l'échéance de l'option.
    strike : float
        Le prix d'exercice du put.
    expiry : float
        La date d'expiration du put, en années.
    vol : float
        La volatilité implicite utilisée pour évaluer le put (en pourcentage).
    rate : float
        Le taux d'intérêt sans risque utilisé dans le modèle (en pourcentage).

    Retours
    -------
    float
        Le prix du put selon le modèle de Black-Scholes.
    """
    
    # Convertir volatilité et taux d'intérêt en fractions
    vol /= 100
    rate /= 100
    
    # Calcul des paramètres d1 et d2
    d1 = (log(spot / strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    d2 = d1 - vol * sqrt(expiry - time)
    
    # Calcul du prix du put
    put_price = -spot * norm.cdf(-d1) + strike * exp(-rate * (expiry - time)) * norm.cdf(-d2)
    
    return put_price
```

```python
from math import log, sqrt
from scipy.stats import norm

def BSCall_Gamma(spot, time, strike, expiry, vol, rate):
    """
    Calcule le gamma d'un call selon le modèle de Black-Scholes.

    Paramètres
    ----------
    spot : float
        Le prix spot de l'actif sous-jacent.
    time : float
        Le temps actuel, en années, jusqu'à l'échéance de l'option.
    strike : float
        Le prix d'exercice du call.
    expiry : float
        La date d'expiration du call, en années.
    vol : float
        La volatilité implicite utilisée pour évaluer le call (en pourcentage).
    rate : float
        Le taux d'intérêt sans risque utilisé dans le modèle (en pourcentage).

    Retours
    -------
    float
        Le gamma du call selon le modèle de Black-Scholes.
    """
    
    # Convertir volatilité et taux d'intérêt en fractions
    vol /= 100
    rate /= 100
    
    # Calcul des paramètres d1
    d1 = (log(spot / strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    
    # Calcul du gamma du call
    call_gamma = norm.pdf(d1) / (spot * vol * sqrt(expiry - time))
    
    return call_gamma
```

```python
from math import log, sqrt, exp
from scipy.stats import norm

def BSCall_Theta(spot, time, strike, expiry, vol, rate):
    """
    Calcule le theta d'un call selon le modèle de Black-Scholes, ajusté pour la variation de la valeur de l'option sur une journée en raison du déclin du temps.

    Paramètres
    ----------
    spot : float
        Le prix spot de l'actif sous-jacent.
    time : float
        Le temps actuel, en années, jusqu'à l'échéance de l'option.
    strike : float
        Le prix d'exercice du call.
    expiry : float
        La date d'expiration du call, en années.
    vol : float
        La volatilité implicite utilisée pour évaluer le call (en pourcentage).
    rate : float
        Le taux d'intérêt sans risque utilisé dans le modèle (en pourcentage).

    Retours
    -------
    float
        Le theta du call selon le modèle de Black-Scholes, ajusté pour une journée.
    """
    
    # Convertir volatilité et taux d'intérêt en fractions
    vol /= 100
    rate /= 100
    
    # Calcul des paramètres d1 et d2
    d1 = (log(spot / strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    d2 = d1 - vol * sqrt(expiry - time)
    
    # Calcul du theta du call
    theta = (-spot * vol * norm.pdf(d1) / (2 * sqrt(expiry - time))
             - rate * strike * exp(-rate * (expiry - time)) * norm.cdf(d2))
    
    return theta
```

```python
from math import log, sqrt, exp
from scipy.stats import norm

def BSCall_Vega(spot, time, strike, expiry, vol, rate):
    """
    Calcule le vega d'un call selon le modèle de Black-Scholes.

    Paramètres
    ----------
    spot : float
        Le prix spot de l'actif sous-jacent.
    time : float
        Le temps actuel, en années, jusqu'à l'échéance de l'option.
    strike : float
        Le prix d'exercice du call.
    expiry : float
        La date d'expiration du call, en années.
    vol : float
        La volatilité implicite utilisée pour évaluer le call (en pourcentage).
    rate : float
        Le taux d'intérêt sans risque utilisé dans le modèle (en pourcentage).

    Retours
    -------
    float
        Le vega du call selon le modèle de Black-Scholes.
    """
    
    # Convertir volatilité et taux d'intérêt en fractions
    vol /= 100
    rate /= 100
    
    # Calcul du paramètre d1
    d1 = (log(spot / strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    
    # Calcul du vega du call
    vega = spot * sqrt(expiry - time) * norm.pdf(d1)
    
    return vega
```

```python
def BSPut_Delta(spot, time, strike, expiry, vol, rate):
    """
        Calcule le delta d'une option put selon le modèle de Black-Scholes.

        Paramètres
        ----------
        spot: float
            Le prix au comptant de l'actif sous-jacent.
        time: float
            Le moment où le prix de l'option doit être évalué.
        strike: float
            Le prix d'exercice de l'option put.
        expiry: float
            La date d'expiration de l'option.
        vol: float
            La volatilité implicite à utiliser pour le calcul du prix (en pourcentage).
        rate: float
            Le taux d'intérêt sans risque à utiliser dans le modèle (en pourcentage).

        Retourne
        -------
        float
            Le delta de l'option put selon le modèle de Black-Scholes.
    """

    vol /= 100
    rate /= 100
    d1 = (log(spot/strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    return -norm.cdf(-d1)
```

```python
def BSPut_Gamma(spot, time, strike, expiry, vol, rate):
    """
        Calcule le gamma d'une option put selon le modèle de Black-Scholes.

        Paramètres
        ----------
        spot: float
            Le prix au comptant de l'actif sous-jacent.
        time: float
            Le moment où le prix de l'option doit être évalué.
        strike: float
            Le prix d'exercice de l'option put.
        expiry: float
            La date d'expiration de l'option.
        vol: float
            La volatilité implicite à utiliser pour le calcul du prix (en pourcentage).
        rate: float
            Le taux d'intérêt sans risque à utiliser dans le modèle (en pourcentage).

        Retourne
        -------
        float
            Le gamma de l'option put selon le modèle de Black-Scholes.
    """

    vol /= 100
    rate /= 100
    d1 = (log(spot/strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    return norm.pdf(d1) / (spot * vol * sqrt(expiry - time))
```

```python
def BSPut_Theta(spot, time, strike, expiry, vol, rate):
    """
        Calcule le theta d'une option put selon le modèle de Black-Scholes, ajusté pour le changement de valeur de l'option en une journée
        en raison de l'amortissement temporel.

        Paramètres
        ----------
        spot: float
            Le prix au comptant de l'actif sous-jacent.
        time: float
            Le moment où le prix de l'option doit être évalué.
        strike: float
            Le prix d'exercice de l'option put.
        expiry: float
            La date d'expiration de l'option.
        vol: float
            La volatilité implicite à utiliser pour le calcul du prix (en pourcentage).
        rate: float
            Le taux d'intérêt sans risque à utiliser dans le modèle (en pourcentage).

        Retourne
        -------
        float
            Le theta d'une option put selon le modèle de Black-Scholes, ajusté pour une journée.
    """

    vol /= 100
    rate /= 100
    d1 = (log(spot/strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    d2 = (log(spot/strike) + (rate - vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    theta = ((spot * vol * norm.pdf(d1)) / (2 * sqrt(expiry - time))) - rate * strike * exp(-rate * (expiry - time)) * norm.cdf(-d2)
    return theta
```

```python
def BSPut_Vega(spot, time, strike, expiry, vol, rate):
    """
        Calcule le vega d'une option put selon le modèle de Black-Scholes.

        Paramètres
        ----------
        spot: float
            Le prix au comptant de l'actif sous-jacent.
        time: float
            Le moment où le prix de l'option doit être évalué.
        strike: float
            Le prix d'exercice de l'option put.
        expiry: float
            La date d'expiration de l'option.
        vol: float
            La volatilité implicite à utiliser pour le calcul du prix (en pourcentage).
        rate: float
            Le taux d'intérêt sans risque à utiliser dans le modèle (en pourcentage).

        Retourne
        -------
        float
            Le vega d'une option put selon le modèle de Black-Scholes.
    """

    vol /= 100
    rate /= 100
    d1 = (log(spot/strike) + (rate + vol**2 / 2) * (expiry - time)) / (vol * sqrt(expiry - time))
    return spot * sqrt(expiry - time) * norm.pdf(d1)
```

# Calcul du prix de l'option Call et le put à l'aide des méthodes de Monte Carlo

Le prix de l’actif satisfait donc l’équation différentielle stochastique suivante :

$$ dS_t = \mu S_t dt + \sigma S_t dW_t $$

où \( $S_t$ \) est le prix de l’actif à l’instant \( $t $\), \( $\mu $\) est le taux de rendement moyen, \( $\sigma \ $) est la volatilité des rendements, et \( $W_t$ \) est un processus de Wiener (c’est-à-dire un mouvement brownien).

où nous avons posé $\mu = r$, puisque nous travaillons dans un cadre du proba risque neutre . Cette équation peut être discrétisée par un schéma d'Euler. En faisant cela, nous obtenons le niveau de l'indice boursier $S_{T}$ à un futur temps $T$ donné un niveau $S_{t}$ à un temps $t$ :
$$
S_{T} = S_{t}\exp\left(\left(r-\frac{1}{2}\sigma^{2}\right)(T-t) +\sigma \sqrt{T-t}\,z\right)\,,
$$
où $z$ est une variable stochastique suivant une distribution normale.


Nous pouvons simuler le processus pour $ S_{T} $ en utilisant la méthode de Monte Carlo. Par exemple, nous définissons $ r=0.06 $ $ \sigma = 0.3 $,$ T-t = 1 $ $ S_{t} = 100 $ Nous effectuons $ I = 100000 $ itérations et choisissons $ M = 100 $ intervalles pour la discrétisation de l'intervalle de temps.


```python
r=0.06
sigma = 0.3
T = 1.0
S0 = 100
K=110
iterations = 1000000
t_intervals = 100


delta_t = T / t_intervals  
Z = np.random.standard_normal((t_intervals + 1, iterations))  
S = np.zeros_like(Z) 
```

```python
S[0] = S0 
for t in range(1, t_intervals + 1):
    S[t] = S[t-1] * np.exp((r - 0.5 * sigma ** 2) * delta_t + sigma * delta_t ** 0.5 * Z[t])
```

### Payoff de l'option Call :
Le payoff d'une option call à l'échéance est donné par :
$ \text{Payoff Call} = \max(S_T - K, 0) $
où :
- \($ S_T $\) est le prix de l'actif sous-jacent à l'échéance,
- \($ K $\) est le prix d'exercice (strike).

### Payoff de l'option Put :
Le payoff d'une option put à l'échéance est donné par :
\[ $\text{Payoff Put} = \max(K - S_T, 0) $\]
où :
- \( $S_T $\) est le prix de l'actif sous-jacent à l'échéance,
- \( $K $\) est le prix d'exercice (strike).

```python
payoff_Call = np.maximum(S[-1] - K, 0)
payoff_Put = np.maximum(K - S[-1] , 0)
```

La  Valeur  de  l'option :
$$
 C=\frac{exp(-r \cdot T) \cdot \sum{p_i}}{iterations}
$$

```python
Mc_call_price = np.exp(-r * T) * np.sum(payoff_Call) / iterations
Mc_call_price
```

```python
Mc_put_price = np.exp(-r * T) * np.sum(payoff_Put) / iterations
Mc_put_price
```
